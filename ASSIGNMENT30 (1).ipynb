```python
import numpy as np
from sklearn import cluster, datasets
import scipy.misc
import matplotlib.pyplot as plt
```

```python
image = scipy.misc.face(gray=False) #-- Open native instead of grayscale so the image has all three dimensions
```

```python
image.shape
```

```python
plt.figure(figsize=(7.68,10.24), dpi=100)
plt.imshow(image)
```

```python
image_reshape = (image / 255).reshape(-1,3)
```

```python
image_reshape.shape
```

```python
from sklearn.cluster import KMeans
k_colors = KMeans(n_clusters=5).fit(image_reshape)
y_pred=k_colors.predict(image_reshape)
```

```python
k_colors.cluster_centers_.shape
```

```python
k_colors.labels_.shape
```

```python
newimg=k_colors.cluster_centers_[k_colors.labels_]
```

```python
newimg.shape
```

```python
newimg=np.reshape(newimg, (image.shape))
```

```python
newimg.shape
```

```python
fig = plt.figure(figsize=(20,20))
ax=fig.add_subplot(1,2,1,xticks=[],yticks=[],title='Original Image')
ax.imshow(image)
ax=fig.add_subplot(1,2,2,xticks=[],yticks=[],title='Color Compressed Image using K-Means with 5 clusters')
ax.imshow(newimg)
plt.show()
```
